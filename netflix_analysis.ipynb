```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
df= pd.read_csv("netflix_titles.csv")

df = df.dropna(subset=['country', 'rating'])
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(),errors='coerce')
sns.countplot(data=df,x='type')
plt.title("Movies vs TV Shows", color='darkblue')
plt.show()
```

```python
genres = df['listed_in'].str.split(',').explode()
genres.value_counts().head(10).plot(kind='bar', colormap='viridis')

plt.title("Top Genres",color='black')
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top Countries")
plt.show()
```

```python
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot()
plt.title("Content Over Years")
plt.show()
```

```python
df['rating'].value_counts().plot(kind='bar', color='orange')
plt.title("Ratings Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```
